## Data Visualization  Using Matplotlib's Object Oriented Method

```python
# Importing the required libraries
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd


# First thing we have to do is upload our dataset
heart_disease = pd.read_csv('heart-disease.csv')
heart_disease

# Before we get started, let us take a sample of our data (We are going to use this sample for our visualization)
over_50 = heart_disease[heart_disease['age'] > 50]
over_50.head()

# Now let us upload our dataset
heart_disease = pd.read_csv('heart-disease.csv')
heart_disease

# Now we have our data uploaded, let us create subplots using the object oriented  method
fig, (ax0, ax1) = plt.subplots(nrows = 2, ncols = 1, figsize =(10, 10), sharex = True)

# Let us now add data to subplot ax0
scatter = ax0.scatter(x = over_50['age'], 
                      y = over_50['chol'], 
                      c = over_50['target'], 
                      cmap = 'winter')

# After adding our data, we can go ahead and customize our data 
ax0.set(title = 'Heart Disease and Cholestrol Levels', ylabel = 'Cholestrol');

# Next, we adjust the subplot ax0 x-axis limits
ax0.set_xlim([50, 80])

# We can now add a legend to ax0
ax0.legend(*scatter.legend_elements(), title = 'target')

# We are going to now add a horizontal line (meanline) to ax0
ax0.axhline(over_50['chol'].mean(), linestyle = '--',color = '#4b0082')

# Let us now create a visualization of our data on ax1 like we did for ax0

# First we add our data to ax1
scatter = ax1.scatter(x = over_50['age'], 
                      y = over_50['thalach'], 
                      c = over_50['target'], 
                      cmap = 'winter')

# Next, we customize ax1
ax1.set(title = 'Heart Disease and Max Heart Rate', xlabel = 'Age', ylabel = 'Max Heart Rate')

# Now let us adjust ax1 y-axis limits so that all of our data can be plotted on our subplots
ax1.set_ylim([60, 200])

# Adjusting ax1 x-axis limits
# ax1.set_xlim([50, 80])Since we have already done this for the ax0, it is not necessary to repeat this step so let us just comment it out

# Now let us add a legend to ax1
ax1.legend(*scatter.legend_elements(), title = 'target')

# Next we add a horizontal line (meanline) to ax1
ax1.axhline(over_50['thalach'].mean(), linestyle = '--', color = '#4b0082')

# Finally, we are going to add a title to our figure
fig.suptitle('Heart Disease Analysis', fontsize = 16, fontweight = 'bold');
```
